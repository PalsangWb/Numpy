```python
#QUESTION 2 (DM_H3)
import matplotlib.pyplot as plt
from matplotlib_venn import venn3

# Define three sample sets
A = {1, 2, 3, 4, 5, 6}
B = {5, 6, 7, 8, 9}
C = {1, 3, 5, 7, 9, 11}

# Individual set sizes
card_A = len(A)
card_B = len(B)
card_C = len(C)

# Pairwise intersections
card_A_B = len(A.intersection(B))
card_A_C = len(A.intersection(C))
card_B_C = len(B.intersection(C))

# Three-way intersection
card_A_B_C = len(A.intersection(B).intersection(C))

# Calculate the size of the unique regions
a_only = card_A - card_A_B - card_A_C + card_A_B_C
b_only = card_B - card_A_B - card_B_C + card_A_B_C
c_only = card_C - card_A_C - card_B_C + card_A_B_C
ab_only = card_A_B - card_A_B_C
ac_only = card_A_C - card_A_B_C
bc_only = card_B_C - card_A_B_C
abc = card_A_B_C

# Visualize with Venn Diagram
plt.figure(figsize=(8, 8))
venn3(subsets=(a_only, b_only, ab_only, c_only, ac_only, bc_only, abc),
      set_labels=('A', 'B', 'C'))
plt.title("Inclusion-Exclusion Principle Visualization")
plt.show()

# Verify the result
# Calculate the union directly
cardinality_union_direct = len(A.union(B.union(C)))

# Apply the inclusion-exclusion formula
cardinality_union_formula = (
    card_A + card_B + card_C -
    card_A_B - card_A_C - card_B_C +
    card_A_B_C
)

# Print and compare the results
print(f"Total elements calculated directly: {cardinality_union_direct}")
print(f"Total elements via inclusion-exclusion formula: {cardinality_union_formula}")

print("\nVerification: Both calculations result in the same total count, confirming the principle.")
```
